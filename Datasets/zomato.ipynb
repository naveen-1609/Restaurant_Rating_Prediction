```python
#df.rates.str.replace('-',np.NaN)
df['rate']=df['rate'].str.replace(" ","")
df['rate']=df['rate'].str.replace('-','NaN')
df['rate']=df['rate'].fillna('NaN')
df['rate'].unique()


df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(",","")
df['approx_cost(for two people)'].unique()
```

```python
df.isna().sum()
```

```python
df=df.rename(columns={"approx_cost(for two people)" : "avg_cost",
                      "listed_in(type)" : "meal_type", 
                      "listed_in(city)" : "city"})
df.head()
```

```python
df.dropna(subset=['rate','avg_cost','rest_type'],inplace=True)
df['avg_cost']=df['avg_cost'].astype(int)
df.shape
```

```python
  ## Removing '/5' from rates
df = df.loc[df.rate != 'NEW'] #it skips if there is no rating for the restaurant
df = df.loc[df.rate != '-'].reset_index(drop=True) # it deletes the index instead of inserting it back
remove_slash5 = lambda x: x.replace('/5','') if type(x) == np.str else x
df.rate = df.rate.apply(remove_slash5).str.strip().astype('float') #applying lambda function onto the data

df['rate']=df['rate'].astype(float)
df.isna().sum()
```

```python
df=df[df.cuisines.isna()==False]
df=df[df.rest_type.isna()==False]
df.isna().sum()
```
